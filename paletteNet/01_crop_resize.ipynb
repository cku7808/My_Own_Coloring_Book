```python
srcimgfolder="./datasets/img/"
refinedfolder="./datasets/refined_img/"
```

```python
import cv2
import os
import pathlib


input_folder = srcimgfolder
output_folder = refinedfolder

# 입력 폴더 내 모든 이미지 파일에 대해 반복
for image_path in pathlib.Path(input_folder).glob("*.jpg"):

    image_file = str(image_path)
    image = cv2.imread(image_file)
    height, width = image.shape[:2]

    # 상하좌우 10% 크롭
    crop_top = int(height * 0.1)
    crop_bottom = int(height * 0.9)
    crop_left = int(width * 0.1)
    crop_right = int(width * 0.9)
    cropped_image = image[crop_top:crop_bottom, crop_left:crop_right]

    # 400x400 리사이즈
    resized_image = cv2.resize(cropped_image, (400, 400))

    #저장
    output_file = os.path.join(output_folder, image_path.name)
    cv2.imwrite(output_file, resized_image)
```
